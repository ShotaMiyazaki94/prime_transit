```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
sns.set(font="times",style="ticks",font_scale=1.6)
from zero_point import zpt
zpt.load_tables()
from astropy import units as u
from astropy.coordinates import SkyCoord
```

```python
PRIME_SKY = np.load("../files/obs_fraction.npy")
catalog = pd.read_csv("../files/target_2000pc.csv")
catalog.columns
```

```python
l = np.where(PRIME_SKY[0]<180,PRIME_SKY[0],PRIME_SKY[0]-360)
b = PRIME_SKY[1]
lc = catalog.GLON
bc = catalog.GLAT
```

```python
plt.subplots(figsize=(14,6))
sc = plt.scatter(l,b,c=PRIME_SKY[2],marker=",",s=80,alpha=1,cmap="viridis")
cb = plt.colorbar(sc,pad=0.01)
cb.set_label("(observable time)/(all night time)")
#cb.set_label("time(target>20$^\circ$|GB<20$^\circ$ and Sun<-10$^\circ$)/time(Sun<-10$^\circ$)",fontsize=15)
sc = plt.scatter(lc,bc,s=catalog.nbstars07,edgecolor="k",color="white",alpha=0.5)
plt.legend(*sc.legend_elements("sizes", num=3),title="Number with P~0.7",loc="upper right")
plt.xlim(-180,175)
plt.ylim(-90,85)
plt.xlabel("Galactic longitude (deg)")
plt.ylabel("Galactic latitude (deg)")
plt.title("stellar clusters and PRIME obs-time fraction during off-bulge season")
```
